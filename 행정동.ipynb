## [3] 주소 형태소 분석

```python
df
```

```python
address_col = df.columns[17]
addr = df[address_col].astype("string")

paren = addr.str.extract(r"\(([^)]*)\)", expand=False)

pattern = r"([가-힣][가-힣0-9-]*[동면])(?=\s|$)"

in_paren = (
    paren
    .str.extract(pattern, expand=False)
    .str.replace(r"[^가-힣]", "", regex=True)
)

addr_outside_paren = addr.str.replace(r"\([^)]*\)", "", regex=True)
in_addr = (
    addr_outside_paren
    .str.extract(pattern, expand=False)
    .str.replace(r"[^가-힣]", "", regex=True)
)

df["행정동"] = in_paren.fillna(in_addr)
```
